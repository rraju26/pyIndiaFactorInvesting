## Analyzing Returns using Asset Pricing Models

### Introduction

Asset Pricing Models (APM) are helpful tools in finance that help determine the value of assets. They play a significant role in everyday investment decisions by assessing the expected returns based on the risks involved. Rational behaviour suggests that an investor earns a return (or premium) for taking risks, and this premium should be higher than the risk-free rate, which is the return earned from an investment with no risk (e.g., short term government bonds). APMs focus on the trade-off between risk and return to determine an asset's worth.

At a very simple level, popular APMs rely on three key assumptions:
1. **Linearity**: The relationship between variables is linear.
2. **Perfect information**: All investors have exactly the same information, and not only do they have the same information, they all have the right information. There is no information asymmetry between market participants.
3. **Efficient markets**: All information is incorporated and reflected in the prices of securities as soon as it becomes available.

While these assumptions appear very theoretical, they do cover a lot of what happens in the real-world, especially when some allowance is made for the assumptions of perfect information and efficient markets. Specifically, information becomes common knowledge *over time* or diffuses over a period, and markets are therefore semi-efficient. All models are simplified abstractions of the real-world, and useful to improve our understanding of the complexity. They rarely claim to explain or predict the real-world markets.

The assumption of linearity allows us to use a simple, intuitive, and powerful statistical technique called regression. 

### Regression

Regression is a statistical method used to model and analyse the relationships between a dependent variable and one or more independent variables. It plots a line that *best fits* the observed values, minimising the sum of squared differences for the sample. The fit's quality is measured by how close this line is to the observed data points, with a perfect fit having a measure of 1 and no relationship having a measure of 0.

#### Types of Regression

- **Simple Linear Regression**: Involves one independent variable and one dependent variable modelled through a linear relationship.
- **Multivariate Regression**: Involves multiple independent variables influencing the dependent variable.

#### Basic Formula of Linear Regression

\[Y = \alpha + \beta X + \epsilon \]

Where:
- \( Y \): Dependent variable (what you are predicting).
- \( X \): Independent variable (predictor).
- \( \alpha \): Intercept of the regression line.
- \( \beta \): Slope of the regression line.
- \( \epsilon \): Error term.

*Alpha* (\(\alpha\)) is the intercept term, representing the value of \( Y \) when \( X \) is zero. *Beta* (\(\beta\)) is the slope of the line, indicating the change in \( Y \) for a one-unit change in \( X \). The error term (\(\epsilon\)) accounts for the differences between the observed and predicted values.

### Capital Asset Pricing Model (CAPM)

The Capital Asset Pricing Model (CAPM), introduced by William Sharpe, Jack Treynors and others in the 1960s following Harry Markowitz's seminal work, explains the relationship between systematic risk and expected return for assets. The formula is given by:

\[ r - r_f = \alpha + \beta_M (MKT - r_f) \]

Where:
- \( r \): Return of the asset.
- \( r_f \): Return of the risk-free asset.
- \( MKT \): Return of the market.

CAPM is widely used for its simplicity and intuitive appeal. However, it assumes a single-period time horizon and homogeneous expectations among market participants. Using CAPM, an investor can estimate the expected return of a stock by considering the stock's beta (its sensitivity to market movements), the risk-free rate, and the market return.

### Multivariate Regression

Multivariate regression, is similar to a simple regression, but involves multiple independent variables influencing the dependent variable. The regression attempts to find the best fit amongst the various predictor variables. Multivariate regression allows us to incorporate additional factors beyond the market return to explain asset returns. While this adds complexity, it can help decompose the effect of various factors.

#### Equation

\[ Y = \alpha + \beta_1 X_1 + \beta_2 X_2 + ... + \beta_n X_n + \epsilon \]

Where:
- \( Y \): Dependent variable (what you are predicting).
- \( X_1, X_2, ..., X_n \): Independent variables (predictors).
- \( \alpha \): Intercept of the regression (when all predictor variables are zero).
- \( \beta_1, \beta_2, ..., \beta_n \): The coefficients for each predictor.
- \( \epsilon \): Error term.

#### Interpretation of Coefficients

- *Alpha* (\(\alpha\)): Represents the value of \( Y \) when all \( X \) variables are zero.
- *Beta* (\(\beta\)): Indicates the change in the dependent variable for a one-unit change in the corresponding independent variable, holding all other variables constant.

#### Assumptions of Multivariate Regression

The multivariate regression has some additional assumptions in the independent variables:
1. **Independence**: The observations of the independent variables are independent of each other.
3. **Homoscedasticity**: The variance of error terms is constant across all levels of the independent variables.
4. **Normality**: The residuals (errors) of the model are normally distributed.

#### Multicollinearity and Autocorrelation

In multivariate regressions, two common issues that can lead to erroneous conclusions if not addressed properly are multicollinearity and autocorrelation.

Multicollinearity occurs when two or more independent variables in a regression model are highly correlated, which can distort the estimated coefficients and make them unreliable. For instance, if we are trying to predict house prices using both the size of the house and the number of rooms, these two variables might be highly correlated because larger houses tend to have more rooms. This can cause difficulty in determining the individual effect of each predictor on the house price. Additionally, the coefficient estimates can become unstable, meaning they can change significantly with small modifications in the model. To detect multicollinearity, we often use the Variance Inflation Factor (VIF); a VIF value greater than 10 typically indicates significant multicollinearity. To address this, one might remove or combine highly correlated variables or use techniques like Principal Component Analysis (PCA) to reduce multicollinearity.

Autocorrelation, on the other hand, occurs when the residuals (errors) are correlated across time periods, which is common in time-series data where one period's outcome influences the next. For example, in predicting stock prices, today's stock price might be influenced by yesterday's price. This violates the assumption of independence and can lead to unreliable estimates, making the model seem more accurate than it actually is, which can result in misleading conclusions. The Durbin-Watson test is commonly used to detect autocorrelation, where a test statistic close to 2 indicates no autocorrelation. To correct for autocorrelation, we can adjust the standard errors to account for the lack of independence or use time-series models such as ARIMA that specifically account for autocorrelation.

### Fama-French 3-Factor Model

The Fama-French 3-factor model (FF3), introduced by Eugene Fama and Kenneth French in 1993, extends CAPM by adding size and value factors, thereby making it a multivariate regression:

\[ r - r_f = \alpha + \beta_M (MKT - r_f) + \beta_S SMB + \beta_V HML \]

Where:
- \( SMB \): Small Minus Big (size).
- \( HML \): High Minus Low (value).

The model addresses some limitations of CAPM by considering the empirical observations that small-cap stocks tend to outperform large-cap stocks, and value stocks (with high book-to-market ratios) tend to outperform growth stocks.

Fama-French solved the inherent multicollinearity of having three portfolios of stocks, all of which will have exposure to the market factor, by designing the size and value portfolios as market-neutral long-short portfolios. The factors in the Fama-French models are constructed to be orthogonal, meaning they are designed to be uncorrelated with each other as much as possible, which helps avoid multicollinearity issues. 

### Carhart 4-Factor Model (Including Momentum)

The 4-factor model further adds a momentum factor to the FF3 model. The momentum factor was proposed by Mark Carhart in 1997, recognising that stocks that have performed well in the past tend to continue performing well in the future:

\[ r - r_f = \alpha + \beta_M (MKT - r_f) + \beta_s SMB + \beta_s HML + \beta_m WML \]

Where:
- \( WML \): Winner Minus Loser (momentum), representing the tendency of stocks that have performed well in the past to continue performing well.

### Fama-French 5-Factor Model

The Fama-French 5-factor model (FF5), introduced by Fama and French in 2015, further adds profitability and investment factors to FF3:

\[ r - r_f = \alpha + \beta_M (MKT - r_f) + \beta_S SMB + \beta_V HML + \beta_R RMW + \beta_C CMA \]

Where:
- \( RMW \): Robust Minus Weak (profitability).
- \( CMA \): Conservative Minus Aggressive (investment).

This model aims to capture more dimensions of risk and return by including factors related to a company's profitability and investment behaviour. To address multicollinearity, the construction of the \(SMB)\ portfolio in FF5 is different compared to FF3, ensuring the factors are as orthogonal as possible.

#### Constructing Factor Portfolios

Fama-French constructed factor portfolios by sorting stocks into 2x3 different groups based on their characteristics (e.g., value, profitability, investment) and size (defined by market cap, into big or small). Each charactertistic had breakpoints (30th and 70th percentiles of big stocks), creating 3 categories of high, neutral, and low characteristics. From here, the market-neutral factor would take long positions in stocks that exhibit the desired characteristics and short positions in stocks that do not, all weighted by market cap, rebalanced annually. This construction ensures that the factor portfolios are market neutral and capture the pure effect of the factor and neatly sidestepping the obvious multicollinearity issue.

## Tests of Significance in Regression

### Introduction to Tests of Significance

One significant advantage of a regression is determining whether the relationships observed between the dependent and independent variables are statistically significant. This involves performing tests of significance, that assess if the results obtained from the regression are likely to be due to chance or if they reflect a true relationship.

### Importance of Tests of Significance

Tests of significance provide a way to quantify the confidence we can have in our regression results. By testing the significance of the regression coefficients, we can determine whether the independent variables have a meaningful impact on the dependent variable. This is useful for making informed decisions based on the model's predictions.

### Common Tests of Significance

#### t-Test for Regression Coefficients

The t-test is used to assess the significance of individual regression coefficients. It evaluates whether each coefficient (\(\beta_i\)) is significantly different from zero. If the t-test indicates that a coefficient is significantly different from zero, it suggests that the corresponding independent variable has a significant impact on the dependent variable.

- **Null Hypothesis (H0)**: The coefficient (\(\beta_i\)) is equal to zero (no effect).
- **Alternative Hypothesis (H1)**: The coefficient (\(\beta_i\)) is not equal to zero (significant effect).

The t-test produces a t-statistic and a p-value. If the p-value is less than a specified significance level (e.g., 0.05), we reject the null hypothesis and conclude that the coefficient is significant. Traditionally, \(t)\-statistics greater than 2 were considered adequate, though practitioners and academics now look for values in excess of 3.5 or higher depending on the test circumstances.

#### F-Test for Overall Regression Model

The F-test assesses the overall significance of the regression model. It tests whether at least one of the regression coefficients is significantly different from zero. This helps us determine if the independent variables, as a group, are good predictors of the dependent variable.

- **Null Hypothesis (H0)**: All regression coefficients are equal to zero (no effect).
- **Alternative Hypothesis (H1)**: At least one regression coefficient is not equal to zero (significant effect).

The F-test produces an F-statistic and a p-value. If the p-value is less than a specified significance level (e.g., 0.05), we reject the null hypothesis and conclude that the regression model is significant.

### Advantages of Regression Analysis

#### Confidence in Conclusions

Regression analysis provides a structured way to evaluate the relationships between variables. By using tests of significance, we can quantify the confidence we have in our conclusions. This helps ensure that the observed relationships are not due to random chance, but rather reflect true underlying patterns.

#### Identifying Key Predictors

Through significance testing, regression analysis helps identify which independent variables are most important in predicting the dependent variable. This is valuable for focusing on the most influential factors and making more accurate predictions.

#### Quantifying Relationships

Regression coefficients quantify the strength and direction of the relationships between variables. This allows for precise measurement of how changes in independent variables affect the dependent variable, providing actionable insights for decision-making.

#### Evaluating Model Fit

Regression analysis includes various metrics, such as R-squared and adjusted R-squared, to evaluate the fit of the model. These metrics indicate how well the model explains the variation in the dependent variable, helping assess its predictive power.

### Summary

In this section, we introduced the concept of tests of significance in regression analysis and their importance in validating the relationships between variables. By using t-tests and F-tests, we can assess the significance of individual coefficients and the overall regression model, respectively. Regression analysis provides confidence in our conclusions, identifies key predictors, quantifies relationships, and evaluates model fit, making it a powerful tool for financial analysis and decision-making.

Lets us take a few Nifty Indices and explore how APM allow for an understanding of the drivers of returns.

```python
# Import Libraries
import pandas as pd
import statsmodels.formula.api as smf
from indiafactorlibrary import IndiaFactorLibrary
```

## Initialise IndiaFactorLibrary and Read Fama-French 6 Factors Dataset
We will initialise the `IndiaFactorLibrary` and read the Fama-French 6 factors dataset, which will be used in our regression analysis. The FF6 data is from Invespar "Data Library: Fama-French Factors, Momentum, and Low-Risk Factors for the Indian Market" available at https://www.invespar.com/research


```python
# Initialize the IndiaFactorLibrary
ifl = IndiaFactorLibrary()
#Read the Fama-French 3 + Momentum Value-Weight Dataset
ff4 = ifl.read('ff4')[0]
# Read the Fama-French 5 + Momentum Value-Weight Dataset
ff6 = ifl.read('ff6')[0]
print(ff6)
```

## Fetch and Read the Example File of Nifty Indices into a DataFrame

Read the daily NAVs into a pandas DataFrame and convert them to monthly returns. The data is sourced from ABSL AMC ([Additional Data on Quant Model ](https://mutualfund.adityabirlacapital.com/market-update)).


```python
# URL of the CSV file on GitHub
url = 'https://raw.githubusercontent.com/rraju26/pyIndiaFactorInvesting/main/data/nifty_example.csv'
# Local file path
file_path = 'data/nifty_example.csv'

# Try to read the CSV file from the local directory, fallback to URL if it fails
try:
    nifty_navs = pd.read_csv(file_path, index_col='Dates', parse_dates=['Dates'])
except FileNotFoundError:
    nifty_navs = pd.read_csv(url, index_col='Dates', parse_dates=['Dates'])

print(nifty_navs.tail())
```

```python
# Resample Monthly  
nifty_navs_monthly = nifty_navs.resample('M').last()

# Calculate Monthly Returns
nifty_rets_monthly = nifty_navs_monthly.pct_change().dropna(how='all') * 100

# Excess Returns 
# Nifty and FF have different start and end dates.
nifty_excess = nifty_rets_monthly.sub(ff6['RF'], axis=0).dropna(how='all')

print(nifty_excess.head())
```

## Concatenate Excess Returns with Fama-French 6 Factors Dataset
We will concatenate the excess returns with the Fama-French 6 factors dataset.


```python
# Prepare Regression Dataframe as a dict
apm_dict = {}
# capm for market
apm_dict['capm'] =  pd.concat([nifty_excess, ff6[['MF']]], axis=1).dropna()
# ff4 for ff3+momentum
apm_dict['ff4'] = pd.concat([nifty_excess, ff4[['MF', 'SMB', 'HML', 'WML']]], axis=1).dropna()
# ff6 for FF5 + momentum 
apm_dict['ff6'] = pd.concat([nifty_excess, ff6[['MF', 'SMB5', 'HML', 'RMW', 'CMA', 'WML']]], axis=1).dropna()
print(apm_dict)
```

## Nifty 100 
The NIFTY_100 is a benchmark index that represents the top 100 companies based on full market capitalisation, accounting for about 69% of the free float market capitalisation of stocks listed on the NSE as of September 29, 2023. It is rebalanced on a semi-annual basis.

```python
regression = {}
for apm in ['capm', 'ff4', 'ff6']:
    # Perform linear regression using statsmodels
    if apm == 'capm':
        regression[apm] = smf.ols('NIFTY_100 ~ MF', data=apm_dict[apm]/100).fit()
    elif apm == 'ff4':
        regression[apm] = smf.ols('NIFTY_100 ~ MF + SMB + HML + WML', data=apm_dict[apm] / 100).fit()
    else:
        regression[apm] = smf.ols('NIFTY_100 ~ MF + SMB5 + HML + RMW + CMA + WML', data=apm_dict[apm] / 100).fit()

    print(f'For {apm} APM ->')
    # Print the regression summary
    print(regression[apm].summary())
    print()
    print()
```

 For the Nifty 100, *ex-ante*, we should expect the *beta* to be close to 1 (as it accounts for 70% of market capitalisation on the NSE), and a statistically significant coefficient for \(SMB\). This coefficient should be negative as the Index consists of the largest companies by market capitalisation. The exposures to other factors are uncertain *ex-ante*, as it is not possible to intuit the other characteristics of these large firms.

The regression outputs provide a comparative analysis of three asset pricing models: CAPM, Fama-French 4-Factor Model, and Fama-French 6-Factor Model, using the NIFTY_100 index as the dependent variable. The CAPM model demonstrates a strong fit with an R-squared of 0.942, indicating that approximately 94.2% of the variability in NIFTY_100 returns can be explained by market returns. The market factor (\(MF\)) coefficient of 0.94 is highly significant, highlighting its substantial impact on NIFTY_100 returns.

The Fama-French 4-Factor Model enhances the explanatory power with an R-squared of 0.971, showing that 97.1% of the variability is captured by the model. Alongside the market factor, the size (\(SMB\)) and value (\(HML\)) factors are significant, with negative coefficients of -0.23 and -0.07 respectively. The \(t\)-statistic for \(HML\) is 3.1, below the 3.5 cut-off mentioned earlier in this notebook, indicating its influence is likely weak. The momentum factor (\(WML\)) is not significant at the 0.05 level, suggesting no influence in this context.

The Fama-French 6-Factor Model further improves the model fit slightly, with an R-squared of 0.973, explaining 97.3% of the return variability. Only \(t\)-statistics for the market and size factors exceed 3.5. While the profitability (\(RMW\)) factor shows significance with a \(t\)-statistic less than 3, this is not statistically significant. 

Under no model did the NIFTY_100 show any evidence of outperformance, as indicated by the lack of a significant coefficient for the intercept.

Overall, the results highlight the increased explanatory power of multi-factor models in capturing the nuances of asset returns compared to the single-factor CAPM.



```python
from statsmodels.iolib.summary2 import summary_col
start = nifty_excess.index[0].strftime('%b %Y')
end = nifty_excess.index[-1].strftime('%b %Y')
print(f'APM Regressions: NIFTY 100 ({start} - {end})')
print(summary_col([regression[x] for x in regression.keys()], 
                  float_format='%.3f', 
                  model_names=['CAPM_NIFTY100', 'FF4_NIFTY100',
                               'FF6_NIFTY100'],
                    regressor_order=['Intercept', 'MF', 'SMB', 'SMB5',
                                     'HML', 'RMW','CMA','WML'],
                 stars=True))
```

We use the `summary_col` function from the `statsmodels.iolib.summary2` module to produce a comparative summary of regression outputs for the NIFTY 100 index from January 2013 to April 2024. This summary includes results from three asset pricing models: CAPM, Fama-French 4-Factor Model, and Fama-French 6-Factor Model. The summary table displays the coefficients, standard errors, and significance levels (indicated by stars) for the regressors, including the market factor (\(MF\)), size (\(SMB\) and \(SMB5\)), value (\(HML\)), profitability (\(RMW\)), investment (\(CMA\)), and momentum (\(WML\)) factors. The R-squared and adjusted R-squared values are also included to indicate the goodness of fit for each model.

This summary table provides a concise comparison of the models, making it easier to identify significant factors and compare the overall explanatory power of each model. It is useful when you need a quick overview or when presenting results to an audience. However, for a deeper understanding and detailed analysis, one should look at the detailed regression outputs, which provide additional statistics such as the t-statistics, p-values, and diagnostic tests like the Durbin-Watson statistic. The detailed outputs help to evaluate the model assumptions and ensure the robustness of the regression results.


## Plot Residuals

```python
import matplotlib.pyplot as plt

# List of model names
model_names = ['capm', 'ff4', 'ff6']

# Calculate the residuals for each model and store in a dictionary
residuals_dict = {model: regression[model].resid for model in model_names}

# Create a DataFrame to hold the residuals
residuals_df = pd.DataFrame(residuals_dict)

# Calculate and print mean and variance of residuals for each model
for model in model_names:
    mean_residual = residuals_df[model].mean()
    var_residual = residuals_df[model].var()
    print(f'\n{model} Residuals:')
    print(f'Mean: {mean_residual:.6f}')
    print(f'Variance: {var_residual:.6f}')

# Plot the residuals
plt.figure(figsize=(12, 6))

for model in model_names:
    plt.plot(residuals_df.index, residuals_df[model], label=f'{model.upper()} Residuals')

plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals Comparison: CAPM, FF4, and FF6')
plt.xlabel('Date')
plt.ylabel('Residuals')
plt.legend()

plt.show()
```

The analysis of residuals for the three asset pricing models—CAPM, Fama-French 4-Factor Model, and Fama-French 6-Factor Model—provides insights into their respective fits. The mean of the residuals for all three models is effectively zero, indicating that the models do not systematically overestimate or underestimate the NIFTY 100 returns. However, the variance of the residuals, which measures the dispersion of the residuals around the mean, decreases as we move from the CAPM to the multi-factor models. Specifically, the CAPM residuals have a variance of 0.000126, the Fama-French 4-Factor Model reduces this to 0.000062, and the Fama-French 6-Factor Model further reduces the variance to 0.000058. This decrease in variance demonstrates that the multi-factor models provide a better fit to the data by capturing more of the return variability through additional factors, leading to smaller and more consistent residuals.

However, the variance of the residuals, which measures the dispersion of the residuals around the mean, decreases as we move from the CAPM to the multi-factor models demonstrating that the multi-factor models provide a better fit to the data by capturing more of the return variability through additional factors, leading to smaller and more consistent residuals.


## NIFTY200 MOMENTUM 30
Finally, we look at the popular Nifty 200 Momentum 30 through the same lens.

```python
regression = {}
for apm in ['capm', 'ff4', 'ff6']:
    # Perform linear regression using statsmodels
    if apm == 'capm':
        regression[apm] = smf.ols('NIFTY200_MOMENTUM30 ~ MF', data=apm_dict[apm]/100).fit()
    elif apm == 'ff4':
        regression[apm] = smf.ols('NIFTY200_MOMENTUM30 ~ MF + SMB + HML + WML', data=apm_dict[apm] / 100).fit()
    else:
        regression[apm] = smf.ols('NIFTY200_MOMENTUM30 ~ MF + SMB5 + HML + RMW + CMA + WML', data=apm_dict[apm] / 100).fit()

print(f'APM Regressions: NIFTY200 MOMENTUM 30 ({start} - {end})')
print(summary_col([regression[x] for x in regression.keys()], 
                  float_format='%.3f', 
                  model_names=['CAPM_NIFTY200_MOM30', 'FF4_NIFTY200_MOM30',
                               'FF6_NIFTY200_MOM30'],
                    regressor_order=['Intercept', 'MF', 'SMB', 'SMB5',
                                     'HML', 'RMW','CMA','WML'],
                 stars=True))
```

### Regression Analysis for NIFTY200 Momentum 30

The NIFTY200 Momentum 30 Index aims to track the performance of 30 high momentum stocks across large and mid-cap stocks. The Momentum Score for each stock is based on recent 6-month and 12-month price returns, adjusted for volatility. The weight of each stock is determined by multiplying the free float market cap with the Normalised Momentum Score, with individual stock weights capped to prevent over-concentration. The index is rebalanced semi-annually in June and December.

The CAPM model demonstrates a good fit with an R-squared of 0.758, indicating that approximately 75.8% of the variability in NIFTY200 Momentum 30 returns can be explained by market returns. The market factor (\(MF\)) coefficient is 0.908, reflecting the substantial impact of market movements on the index returns. The intercept is significantly positive, indicating an *alpha* or outperformance of the index over the market exposure.

The Fama-French 4-Factor Model improves the explanatory power significantly with an R-squared of 0.862. Alongside the market factor, the momentum factor (\(WML\)) is significant with a coefficient of 0.411 (\(t\)-statistic well over 3.5), reflecting the index's strong exposure to momentum stocks, which is expected given the index's design. The size (\(SMB\)) and value (\(HML\)) factors are not significant, indicating that these factors do not have a notable impact on the index returns.

The Fama-French 6-Factor Model further improves the model fit slightly, with an R-squared of 0.867. The market factor (\(MF\)) coefficient increases to 1.050, and the momentum factor (\(WML\)) remains highly significant with a coefficient of 0.398 (\(t\)-statistic again in excess of 3.5). The profitability (\(RMW\)) and investment (\(CMA\)) factors are not significant, suggesting limited influence in this context. The size and value factors remain insignificant.

Overall, the regression results for the NIFTY200 Momentum 30 highlight the dominant role of market and momentum factors in explaining the index returns. The positive and significant intercept in the CAPM model indicates an abnormal return component, while the multi-factor models, particularly the inclusion of the momentum factor, provide better explanatory power for the index's performance. Note that the \(WML\) is less than 1, indicating that the index does not capture the full academic momentum effect, but only about 40%.


### Residuals

```python
# List of model names
model_names = ['capm', 'ff4', 'ff6']

# Calculate the residuals for each model and store in a dictionary
residuals_dict = {model: regression[model].resid for model in model_names}

# Create a DataFrame to hold the residuals
residuals_df = pd.DataFrame(residuals_dict)

# Calculate and print mean and variance of residuals for each model
for model in model_names:
    var_residual = residuals_df[model].var()
    print(f'\n{model} Residuals:')
    print(f'Mean: {mean_residual:.6f}')
    print(f'Variance: {var_residual:.6f}')

# Plot the residuals
plt.figure(figsize=(12, 6))

for model in model_names:
    plt.plot(residuals_df.index, residuals_df[model], label=f'{model.upper()} Residuals')

plt.axhline(0, color='red', linestyle='--')
plt.title('NIFTY200 MOMENTUM 30 Residuals Comparison: CAPM, FF4, and FF6')
plt.xlabel('Date')
plt.ylabel('Residuals')
plt.legend()

plt.show()
```

The analysis of residuals for the three asset pricing models is similar to the others we have seen: the mean of the residuals for all three models is effectively zero, indicating that the models do not systematically overestimate or underestimate the returns while the variance of the residuals decreases as we move from the CAPM to the multi-factor models demonstrating that the multi-factor models provide a better fit to the data by capturing more of the return variability through additional factors, leading to smaller and more consistent residuals.
